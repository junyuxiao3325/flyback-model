本程序计算vout到vc的传输函数//
$$
g_{closed}=g_{vc}\frac{R_{FBB}}{R_{FBT}+R_{FBB}}g_{ea}g_{OPTO}\\
g_{ea}=g_m\frac{1+sR_{COMP}C_{COMP}}{sC_{COMP}}\\
g_{OPTO}=\frac{CTRR_{OPTO}}{R_LED}\frac{1}{1+sR_{OPTO}C_{OPTO}}
$$

```python
rfbt=90e3,rfbb=10e3,gm=24e-6,rcomp=70e3,ccomp=2.2e-9,ctr=0.25,ropto=22e3,rled=1e3,copto=2.2e-9
gea=gm*(1+s*rcomp*ccomp)/s/ccomp
gopto=ctr*ropto/rled/(1+s*ropto*copto)
gclosed=gvc*rfbb/(rfbt+rfbb)*gea*gopto
```
